# Net Promoter Score:
https://medium.com/doter/what-is-a-good-nps-score-2021-benchmark-a40ddf6d4b60

```python
df['nps_n'] = df['nps'].replace({'10 （完全愿意）': 10})
df['nps_n'] = df['nps_n'].replace({'0 （完全不愿意）': 0})
df['nps_n'] = df['nps_n'].astype(int)
df.nps_n.describe()
```

```python
sns.histplot(df.nps_n, discrete=True).set_title('Net Promoter Score', fontsize=14);
```

```python
def nps_bucket(x):
    if x > 8:
        bucket = 'promoter'
    elif x > 6:
        bucket = 'passive'
    elif x>= 0:
        bucket = 'detractor'
    else:
        bucket = 'no score'
    return bucket

# check NPS bucket distribution
df['nps_bucket'] = df['nps_n'].apply(nps_bucket)
df['nps_bucket'].value_counts()
```

```python
# calculate NPS
(df['nps_bucket'].value_counts()['promoter'] - df['nps_bucket'].value_counts()['detractor'])/df['nps_bucket'].count()
```

```python
def plt_single_choice_hue(df, col, title, hue_group='group'):
    g = sns.histplot(data=df[[col, hue_group]].sort_values(by=col), x=col, 
#                      stat='percent',
                     discrete=True, hue=hue_group, multiple="dodge", shrink=.8)
    g.set_title(title, font="Hei", fontsize=14)
    g.set_xticklabels(sorted(df[col].value_counts().index), font="Hei", fontsize=14);
    plt.show()
#     return g
```

```python
def plt_single_choice(df, col, title):
    g = sns.histplot(df[col].sort_values(), discrete=True, stat='percent', shrink=.8)
    g.set_title(title, font="Hei", fontsize=14)
    g.set_xticklabels(sorted(df[col].value_counts().index), font="Hei", fontsize=14);
    plt.show()
#     return g
```

```python
def plt_single_choice2(df, col, title):
    plt_single_choice(df, col, title)
    plt_single_choice_hue(df, col, title,hue_group='group')
```

# Visualize multi-answer question by group (Checkboxes)

```python
# Prepare the summary table (please clean the data beforehand)
def prepare_table(data, column_range, group_column_name='group'):
    res = []
    col_range_index = list(column_range)
    group_index = data.columns.to_list().index(group_column_name)
    col_range_index.append(group_index)
    series = data.iloc[:, col_range_index].groupby([group_column_name]).count().unstack()
    print(series)
    for group in series.index.levels[1]:
        for var in series.index.levels[0]:
            res.append(
                [
                    (series[var][group] / data.loc[:,group_column_name].value_counts()[group] * 100).round(2),
                    var,
                    group,
                ]
            )
    return pd.DataFrame(columns=["percentage(%)", "options", group_column_name], data=res)


# Generate the checkbox chart based on the summary table
def gen_chart_checkbox(data, column_range, group_column_name='group'):
    listOfGroup = list(data.loc[:,group_column_name].unique())
    listOfGroup.sort()
    table_sum = prepare_table(data, column_range, group_column_name)

    print("Number of answers in each group: ")
    print(data.loc[:,group_column_name].value_counts())

    fig, ax = plt.subplots(figsize=(10, 8))
    ax = sns.barplot(
        x="percentage(%)",
        y="options",
        hue=group_column_name,
        hue_order=listOfGroup,
        data=table_sum,
    )
    ax.set(xlim=(0, 100))
    ax.set_ylabel("")
#     plt.title(codebook.iloc[column_range[0], 0], fontsize=15)
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0, title="group")
    return plt.show()
```
